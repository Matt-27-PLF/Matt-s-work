```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/kaggle/input/life-expectancy-who-updated/Life-Expectancy-Data-Updated.csv')
df
```

```python
df.columns
```

```python
sns.scatterplot(x='Alcohol_consumption',hue='Region',y='Adult_mortality',data=df,palette='viridis')
plt.xlabel('alcohol', fontsize=12)
plt.ylabel('adult mortality', fontsize=12)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()
```

```python
sns.scatterplot(x='Life_expectancy',hue='Economy_status_Developing',y='GDP_per_capita',data=df,palette='viridis')
plt.xlabel('life exectancy', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()
```
